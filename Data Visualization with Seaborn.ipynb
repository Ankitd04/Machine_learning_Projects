# Data visualization with Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
df = pd.read_csv("G:\datasets\Automobile.csv")
df.head()
```

```python
df.shape
```

# Plotting Univariate Distributions 

```python
# distplot function 

sns.distplot(df["normalized_losses"])
```

# Histogram

```python
# Histogram

sns.distplot(df["normalized_losses"], kde=False, rug=True)
```

# Plotting Bivarient Distribution

# Scatterplot

```python
# jointplot function 

import scipy.stats as stats

sns.jointplot(df['engine_size'], df['horsepower']).annotate(stats.pearsonr)
```

# Hex bin plot

```python
sns.jointplot(df['engine_size'], df['horsepower'], kind='hex')
```

# Kernel Density Estimation

```python
sns.jointplot(df['engine_size'], df['horsepower'], kind='kde')
```

# Visualizing Pairwise relationships in dataset

```python
# pairplot function 

sns.pairplot(df[['normalized_losses', 'engine_size', 'horsepower']])
```

# Plotting Categorical Data

```python
# stripplot function

sns.stripplot(df['fuel_type'], df['horsepower'], jitter=False)
```

```python
# jitter=True

sns.stripplot(df['fuel_type'], df['horsepower'])
```

```python
# swarmplot function

sns.swarmplot(df['fuel_type'],df['horsepower'])
```

```python
df['fuel_type'].value_counts()
```

# Boxplot 

```python
sns.boxplot(df['number_of_doors'], df['horsepower'], hue=df['fuel_type'])
```

# Barplots

```python
sns.barplot(df['body_style'],  df['horsepower'], hue=df['engine_location'])
```

# Countplot

```python
sns.countplot(df['body_style'])
```

# Pointplot

```python
# Pointplot

sns.pointplot(df['fuel_system'], df['horsepower'], hue=df['number_of_doors'])
```

# Drawing Multi-Panel Categorical Plots

```python
# factorplot/ catplot (new release name) function

sns.catplot(x='fuel_type', y='horsepower',
              hue='number_of_doors',
              col='engine_location',
              kind='violin',
              data = df)
```

# function to draw regression models

```python
sns.lmplot(x='horsepower', y='peak_rpm', hue='fuel_type', data=df)
```

```python
sns.regplot(x='engine_size', y='horsepower', data=df)
```
